```python
#let's open it with pandas
import pandas as pd
from pandas import Series,DataFrame
import numpy as np

# For Visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set up the results csv file as a DataFrame
df = pd.read_csv('results.csv')
```

```python
# Let's see a preview of the data
df.head()
```

```python
# We could also get overall info for the dataset
df.describe()
```

```python
#What was the moving average of the global temperature 
sg_avg = df['city_avg_temp'].rolling(50).mean()
plt.plot(df['year'],sg_avg,label ='Singapore')
plt.legend()
plt.xlabel('year')
plt.ylabel('Temperature (C)')
plt.title('Singapore Average temperature')
plt.show()
```

```python
#combine global temperature and Singapore temperature
glb_avg = df['global_avg_temp'].rolling(50).mean()
sg_avg = df['city_avg_temp'].rolling(50).mean()
plt.plot(df['year'],glb_avg,label ='Global')
plt.plot(df['year'],sg_avg, label = 'Singapore')
plt.legend()
plt.xlabel('year')
plt.ylabel('Temperature (C)')
plt.title('Global-Singapore Average temperature')
plt.show()
```

```python
tech_rest =df.pct_change()
sns.jointplot('city_avg_temp','global_avg_temp',tech_rest,kind='scatter',color='seagreen')
plt.show()
```

```python
#change in temperature
glb_rest = df['global_avg_temp'].pct_change()
glb_rest.plot(figsize=(12,4),legend=True,linestyle='--',marker='o')
sg_rest = df['city_avg_temp'].pct_change()
sg_rest.plot(figsize=(12,4),legend=True,linestyle='--',marker='o')
```
